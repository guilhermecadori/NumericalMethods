```python
# Importing support library
import matplotlib.pyplot as plt

# Implementing Newton-Raphson method
def NewtonRaphsonMethod(f, f_lin, x0, epsilon, iterMax=50):
    # Test if current starting point is a root
    if abs(f(x0)) <= epsilon:
        return False, [x0]
    
    # Lists to store values for plotting
    x_values = [x0]
    fx_values = [f(x0)]
    
    # Table header
    print("k\t x\t\t f(x)\t\t f'(x)")
    print('%d\t%.6f\t%.6f\t%.6f' % (0, x0, f(x0), f_lin(x0)))
    
    # Main loop
    for k in range(1, iterMax + 1):
        # Calculating x1 from x0
        x1 = x0 - f(x0) / f_lin(x0)
        
        # Storing values for plotting
        x_values.append(x1)
        fx_values.append(f(x1))
        
        # Printing results
        print('%d\t%.6f\t%.6f\t%.6f' % (k, x1, f(x1), f_lin(x1)))
        
        # Stopping criteria test
        if abs(f(x1)) <= epsilon:
            return False, x_values
        
        # Updating the value of x0
        x0 = x1
        
    print('Error: maximum allowed iterations reached')
    
    return True, x_values

# Creating a general function to be used
def f(x):
    return x**3 - 9*x + 3

# Implementing the first derivative of the above f(x) function
def f_lin(x):
    return 3 * x**2 - 9  # Returning the derivative value

# Setting parameters for testing purposes
x0 = 20
epsilon = 0.00001
maxIter = 50

errorFound, x_values = NewtonRaphsonMethod(f=f, f_lin=f_lin, x0=x0, epsilon=epsilon, iterMax=maxIter)

# Returning root found
print('\nRoot Found:')
print(round(x_values[-1], 4))

# Plotting f(x) values at each iteration
plt.figure(figsize=(10, 6))
iterations = range(len(x_values))
plt.plot(iterations, [f(x) for x in x_values], marker='o', color='skyblue', label='f(x) values')
plt.xlabel('Iteration')
plt.ylabel('f(x)')
plt.title('f(x) Convergence - Newton-Raphson Method')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()

```
