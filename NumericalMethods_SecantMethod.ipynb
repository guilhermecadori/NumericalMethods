```python
# Importing support library
import matplotlib.pyplot as plt

# Implementing the Secant Method for fiding the roots of a function
def secantMethod(f, x0, x1, epsilon, iterMax=50):
    # Test if initial guesses bracket the root
    if f(x0) * f(x1) > 0:
        print("Error: Initial guesses do not bracket the root.")
        return True, None, None
    
    # Lists to store iteration history
    x_values = [x0, x1]
    fx_values = [f(x0), f(x1)]

    # Table header
    print('k\t x\t\t f(x)')

    # Main loop
    for k in range(1, iterMax + 1):
        # Calculating x2 from x0 and x1
        x2 = (x0 * f(x1) - x1 * f(x0)) / (f(x1) - f(x0))

        # Storing values for plotting
        x_values.append(x2)
        fx_values.append(f(x2))
        
        # Printing iteration values
        print('%d\t%.6f\t%.6f' % (k, x2, f(x2)))
        
        # Stopping criteria test
        if abs(f(x2)) <= epsilon:
            return False, x2, x_values, fx_values
        
        # Updating values for the next iteration
        x0 = x1
        x1 = x2
        
    print('Error: Maximum allowed iterations reached.')
    return True, x2, x_values, fx_values

# Creating a general function to be used
def f(x):
    return x**3 - 9*x + 3

# Setting parameters for testing purposes
x0 = -2
x1 = 2
epsilon = 0.00001
maxIter = 50

errorFound, root, x_values, fx_values = secantMethod(f=f, x0=x0, x1=x1, epsilon=epsilon, iterMax=maxIter)

# Returning root found
if not errorFound:
    print('\nRoot Found:')
    print(round(root, 4))
    
    # Plotting f(x) values at each iteration
    plt.figure(figsize=(10, 6))
    iterations = range(len(x_values))
    plt.plot(iterations, fx_values, marker='o', color='skyblue', label='f(x) values')
    plt.xlabel('Iteration')
    plt.ylabel('f(x)')
    plt.title('f(x) Convergence - Secant Method')
    plt.legend()
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.show()

    
```
